```python
import pymc as pm
import matplotlib.pyplot as plt
import numpy as np
import arviz as az
```

```python
ones = np.ones(7)
zeros = np.zeros(3)
occurrences = np.concatenate((ones, zeros))
with pm.Model() as model:
    p = pm.Uniform("p", lower=0, upper=1)
    obs = pm.Bernoulli("obs", p, observed=occurrences)
    idata = pm.sample(2000, tune=2500)
az.plot_trace(idata, show=True, legend=False, figsize=(8, 4))
az.summary(idata)
```

rednia 0.669 z szerokim odchyleniem standardowym 0.131co sugeruje du niepewno co do wartoci 
Przedzia HDI [0.429, 0.902] jest do szeroki, co wiadczy o mniejszej pewnoci co do wartoci prawdopodobiestwa.
